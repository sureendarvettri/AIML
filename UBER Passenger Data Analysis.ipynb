```python
#13. Total no. of trips in each day

df['START_DATE*'] = df['START_DATE*'].str.replace('-','/')
df['END_DATE*'] = df['END_DATE*'].str.replace('-','/')
df['START_DATE*'].value_counts()
```

```python
#14. Average time taken to cover a mile in single trip (distinguish drop, single unknwn and both unknown)
df1 = df[df['START_DATE*'] == df['END_DATE*']]
time = (df1['END_TIME*'].str.replace(':','.').astype(float) - df1['START_TIME*'].str.replace(':','.').astype(float))/df1['MILES*']
time.mean()
```

```python
start = df1['START*'] == 'Unknown Location'
stop = df1['STOP*'] == 'Unknown Location'
df_uk_round = df1[start & stop]
time = (df_uk_round['END_TIME*'].str.replace(':','.').astype(float) - df_uk_round['START_TIME*'].str.replace(':','.').astype(float))/df_uk_round['MILES*']
time.mean()
```

```python
df_uk_single = df1[start | stop]
time = (df_uk_single['END_TIME*'].str.replace(':','.').astype(float) - df_uk_single['START_TIME*'].str.replace(':','.').astype(float))/df_uk_single['MILES*']
time.mean()
```

```python
#15. Average time taken to cover a mile in round trip 
df1 = df[df['START_DATE*'] == df['END_DATE*']]

start = df1['START*'] != 'Unknown Location'
stop = df1['STOP*'] != 'Unknown Location'

df_uk_round = df1[start & stop]
time = (df_uk_round['END_TIME*'].str.replace(':','.').astype(float) - df_uk_round['START_TIME*'].str.replace(':','.').astype(float))/df_uk_round['MILES*']
time.mean()
```
